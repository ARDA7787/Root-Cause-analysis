```python
from tensorflow.keras import utils
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"/Users/nikhildonde/Downloads/RCA_data.csv")
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['ROOT_CAUSE'] = le.fit_transform(df['ROOT_CAUSE'])
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
import tensorflow as tf
```

```python
import keras
```

```python
X = df.drop(['ROOT_CAUSE'], axis = 1)
y = df['ROOT_CAUSE']
y = tf.keras.utils.to_categorical(y,3)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
EPOCHS = 40
BATCH_SIZE = 100
VERBOSE = 1
OUTPUT_CLASSES = len(le.classes_)
```

```python
model = tf.keras.models.Sequential()
model.add(keras.layers.Dense(128, input_shape=(8,), name='Dense-Layer-1', activation='relu'))
model.add(keras.layers.Dense(128, name='Dense-Layer-2', activation='relu'))
model.add(keras.layers.Dense(OUTPUT_CLASSES, name='Final', activation='softmax'))
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, batch_size=BATCH_SIZE, epochs=EPOCHS, verbose=VERBOSE, validation_split=0.2)
```
